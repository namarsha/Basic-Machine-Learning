### Step 1 – Load the data
Load `trip_data.csv`, select the salient features and engineer new features (if necessary).

```python
#TODO: Load the data
file_path   = 'trip_data.csv' #TODO: provide the file path or URL to the dataset `trip_data.csv`

df = pd.read_csv(file_path)

## Create time duration feature
df['lpep_dropoff_datetime_s'] = df['lpep_dropoff_datetime'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S').timestamp())
df['lpep_pickup_datetime_s'] = df['lpep_pickup_datetime'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S').timestamp())
                                                                                                     
df['trip_duration'] = df['lpep_dropoff_datetime_s'] - df['lpep_pickup_datetime_s']  #TODO: store the elapsed time (from above) into the dataframe.

df.head()

#Remove the `lpep_pickup_datetime` and `lpep_dropoff_datetime` from the dataframe.
del df['lpep_dropoff_datetime']
del df['lpep_pickup_datetime']
del df['lpep_dropoff_datetime_s']
del df['lpep_pickup_datetime_s']

#removed features: rateCodeID, PULocationID, DOLocationID

features = ['passenger_count', 'trip_distance', 'fare_amount', 'extra', 'mta_tax',
       'tolls_amount', 'improvement_surcharge', 'total_amount', 'trip_type',
       'congestion_surcharge', 'trip_duration']
X = df[features]
y = df['tip']
```

### Step 3 – Create the Classifier

```python
#Create the RandomForestClassifier

rf_classifier = RandomForestClassifier(criterion= 'gini',  #TODO: provide an option 
                                    max_depth= 15,     #TODO: provide a value. The depth should not exceed 15.
                                    max_features= 7,  #TODO: provide a suitable value
                                    min_samples_split= 8,#TODO: provide a suitable value
                                    n_estimators= 100, #TODO: provide a suitable value
                                    random_state=seed,#TODO: provide a suitable value
                                    bootstrap = False
                                    )

#TODO: Train classifer using the `fit` method
rf_classifier.fit(X_train, y_train)

#TODO: Make predictions with the test set
y_pred = rf_classifier.predict(X_test)

#TODO: evaluate the accuracy
accuracy_score(y_test, y_pred)
```

An ensemble method essentially combines the output of multiple training methods. In scikit-learn, the random forest classifier works on a majority vote system. This vote system makes the models resilient against noise, and thus inductive bias of any one model is reduced, as it is outweighed by the votes from the other, more correct models. As long as those training methods are different, which is ensured by changing the parameters of the trees, the ensemble method is generally much, much more accurate.
