```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the data into a dataframe
data = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/bad-drivers/bad-drivers.csv')

#view samples of the data
data.head()
```

```python
data.describe()

#renaming columns:
data.columns = ['STATE', 'COLLISIONS', 'PERCSPEEDING', 'PERCALCOHOL', 'PERCUNDIST',
               'PERCNOPREV', 'INSURANCEPREM', 'INSURANCELOSS']
data_a = data

data_n = data.drop('STATE', axis=1)

data_n.head()
```

```python
data_n.describe()
```

## Visualize the correlations

```python
#TODO:
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data_n.corr(), square = True, cmap = 'magma', annot = True)
```

For a given state, collisions and insurance premiums are negatively correlated, as are percentages of alcohol induced collisions and percentages of no previous collision. (States where there are a lot of alcohol caused collisions tend to have more collisions that were not the first). States with more distracted drivers tend to have accidents that were not the first. The strongest positive correlation is between insurance premiums and insurance losses.
